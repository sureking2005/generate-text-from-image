```python
#import important packages and libraries
import cv2
import easyocr
import numpy as np
from numpy import ndarray
from google.colab.patches import cv2_imshow #cv2.imshow does not work here
```

```python
#read an Image
image_path = '/content/Screenshot (77).png'
    #the images (license_plate and quotepng) is upload on my Github, you may download from there
    # https://github.com/BushraAijaz/dataset/blob/main/Pakistan_-_License_Plate_-_Sindh%20(1).png
    # https://github.com/BushraAijaz/dataset/blob/main/quotepng.png
    # the image is to uploaded everytime you use this file, as I mentioned in video
img = cv2.imread(image_path)
cv2_imshow(img)
```

```python
#instant text reader
picture_read =easyocr.Reader(['en','ar'],gpu=False) #by default GPU is true, if you don't have GPU, type 'False'
```

```python
#raw text detection on image
picture_results = picture_read.readtext(image_path)
for t in picture_results:
  print (t)
```

```python
#writing text and drawing box on the image

conf_threshold = 0.2
for y in picture_results:
  if y[2]> conf_threshold:
    top_left_corner= [int(value) for value in y[0][0]]
    bottom_right_corner = [int(value) for value in y[0][2]]
    text= y[1]
    img = cv2.rectangle(img, top_left_corner, bottom_right_corner, (0,0,255),2) #dtails of bounding box
    img =cv2.putText(img, text, top_left_corner, cv2.FONT_HERSHEY_PLAIN, 2, (255,0,0), 2) #details of text

cv2_imshow(img)
```
